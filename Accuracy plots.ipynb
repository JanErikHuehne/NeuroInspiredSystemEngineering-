```python
import matplotlib.pyplot as plt
import numpy as np

# change into avg speeds?

# Your data
data = [
    [1, 13, 90, 6.666666667],
    [2, 13, 80, 6.666666667],
    [3, 13, 93.33333333, 4.444444444],
    [1, 26, 93.33333333, 8.888888889],
    [2, 26, 83.33333333, 11.11111111],
    [3, 26, 90, 6.666666667],
    [1, 100, 90, 6.666666667],
    [2, 100, 83.33333333, 4.444444444],
    [3, 100, 80, 6.666666667]
]

# Convert data to a numpy array for easier manipulation
data = np.array(data)

# Extract unique times and persons
times = np.array(['Fast',' Medium','Slow'])
#NOTE SPEED 1 CORRESPONDED TO AND AVG TIME OF,,,, 2 ... AND 3 ... RESPECTIVELY 
persons = np.unique(data[:, 0])

# Plotting
fig, ax = plt.subplots()

bar_width = 0.2
index = np.arange(len(times))

for person in persons:
    person_data = data[data[:, 0] == person]
    avg_accuracy = person_data[:, 2]
    std_dev = person_data[:, 3]

    ax.bar(
        index + bar_width * (person - 1),
        avg_accuracy,
        bar_width,
        label=f'Participant {int(person)}',
        yerr=std_dev,
        capsize=5,
    )

ax.set_xlabel('Detection Speed')
ax.set_ylabel('HSDA in Percentages')
ax.set_title('HSDA vs Detection Speed')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(times)
ax.legend()
#counting up from 0 to 9
plt.show()
```

```python
# 1-way anova for significance testing

import numpy as np
from scipy.stats import f_oneway

# Your data
speeds = [1, 2, 3]
detection_accuracy = [
    [100, 90, 80, 70, 90, 80, 90, 90, 100],
    [80, 100, 90, 80, 90, 80, 80, 70, 90],
    [100, 80, 100, 70, 80, 100, 90, 80, 100],
]

# Perform one-way ANOVA
statistic, p_value = f_oneway(*detection_accuracy)

# Output results
print(f"One-way ANOVA:")
print(f"  F-statistic: {statistic}")
print(f"  p-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The detection accuracy means are significantly different.")
else:
    print("No significant difference in detection accuracy means.")

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_rel

# Your data
data = [
    ["Fast", 87.77777778, 7.407407407],
    ["Slow", 84.44444444, 7.160493827],
    ["Medium", 88.88888889, 10.12345679],
]

# Convert data to a numpy array for easier manipulation
data = np.array(data)

# Extract information
categories = data[:, 0]
average_accuracy = data[:, 1].astype(float)
std_dev = data[:, 2].astype(float)

# Plotting
fig, ax = plt.subplots()

bar_width = 0.5
index = np.arange(len(categories))

# Plot bars
bars = ax.bar(
    index,
    average_accuracy,
    bar_width,
    yerr=std_dev,
    capsize=5,
    color='blue',
)

ax.set_xlabel('Detection Speed')
ax.set_ylabel('HSDA in Percentages')
ax.set_title('Average HSDA vs Detection Speed')
ax.set_xticks(index)
ax.set_xticklabels(categories)
ax.set_ylim(0, 100)  # Set the y-axis limit to 100 for percentage values

# Perform t-tests and add ns annotations
for i in range(len(categories)):
    # Add ns annotation for each bar
    ax.annotate('ns', xy=(index[i] + bar_width / 2, average_accuracy[i]),
                ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your new data
data_new = [
    [1, 81.75824176, 10.02197802],
    [2, 80.22058824, 6.264705882],
    [3, 90.63095238, 5.719047619],
    [4, 95.04397008, 1.756759715],
]

# Convert data to a numpy array for easier manipulation
data_new = np.array(data_new)

# Extract participant numbers
participants_new = data_new[:, 0]

# Calculate the average of all participants
average_all = np.mean(data_new[:, 1])

# Define colors for each participant
participant_colors = ['red', 'green', 'blue', 'purple']

# Plotting
fig, ax = plt.subplots()

bar_width = 0.5
index = np.arange(len(participants_new) + 1)  # +1 for the average bar

# Plot bars for individual participants with different colors
for i, participant in enumerate(participants_new):
    ax.bar(
        index[i],
        data_new[i, 1],
        bar_width,
        yerr=data_new[i, 2],
        capsize=5,
        color=participant_colors[i],
        label=f'Participant {int(participant)}',
    )

# Plot the bar for the average of all participants
ax.bar(
    index[-1],
    average_all,
    bar_width,
    yerr=np.std(data_new[:, 1]),  # Standard deviation of the average
    capsize=5,
    color='orange',
    label='Average over All Participants',
)

ax.set_xlabel('Participant')
ax.set_ylabel('HSDA in Percentages')
ax.set_title('HSDA per Participant')
ax.set_xticks(index)
ax.set_xticklabels(np.concatenate([participants_new, ['Average']]))
ax.set_ylim(0, 100)  # Set the y-axis limit to 100 for percentage values
ax.legend()

plt.show()
```
